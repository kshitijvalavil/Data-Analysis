```python
df=pd.read_csv('/Users/kshitij/Desktop/WidhyaAnalytics/bitcoin_dataset.csv')
#df.fillna(df.median, inplace=True)
df=df.replace([np.inf,-np.inf,np.nan], 0)
df.dtypes
```

```python
df.iloc[1023,1]
```

```python
sns.pairplot(df, x_vars=['btc_market_cap','btc_n_transactions', 'btc_miners_revenue','btc_cost_per_transaction', 'btc_difficulty','btc_hash_rate','btc_cost_per_transaction_percent'], y_vars= 'btc_market_price', size =7,kind ='reg')
plt.show()
```

```python
feature_cols = ['btc_market_cap','btc_n_transactions', 'btc_miners_revenue','btc_cost_per_transaction', 'btc_difficulty','btc_hash_rate','btc_cost_per_transaction_percent']
x= df[feature_cols]
y= df['btc_market_price']
```

```python
x.isnull().sum()
```

```python
np.isinf(x).sum()
```

```python
# Test train split for supervised training
x_train, x_test, y_train, y_test = train_test_split(x, y)
x_test.shape
```

```python
# Create a Linear Model and train it
reg= LinearRegression()
reg.fit(x_train, y_train)
```

```python
print(reg.intercept_)
```

```python
print(reg.coef_)
```

```python
y_pred = reg.predict(x_test)
```

```python
print(metrics.mean_squared_error(y_test, y_pred))
```
